```python
import pandas as pd

master = pd.read_csv("../data/processed/airbnb_master.csv")

print(master.shape)
master.head()
```

```python
master["revenue_category"] = pd.cut(
    master["estimated_revenue_l365d"],
    bins=[-1, 0, 10000, 25000, 20000000],
    labels=["No Revenue", "Low", "Medium", "High"]
)

master["revenue_category"].value_counts()
```

```python
master["price_category"] = pd.cut(
    master["price"],
    bins=[0,100,250,500,1000000],
    labels=["Budget","Standard","Premium","Luxury"]
)

master["price_category"].value_counts()
```

```python
master["host_since"] = pd.to_datetime(
    master["host_since"],
    errors="coerce"
)

master["host_experience_years"] = (
    pd.Timestamp.today() - master["host_since"]
).dt.days / 365

master["host_experience_years"].describe()
```

```python
master["first_review_date"] = pd.to_datetime(
    master["first_review_date"],
    errors="coerce"
)

master["last_review_date"] = pd.to_datetime(
    master["last_review_date"],
    errors="coerce"
)

master["review_lifetime_days"] = (
    master["last_review_date"] -
    master["first_review_date"]
).dt.days

master["review_lifetime_days"].describe()
```

```python
master.to_csv(
    "../data/processed/airbnb_featured.csv",
    index=False
)

print("Feature engineered dataset saved.")
print(master.shape)
```
